```python
df.head()
```

```python
df.columns=["Employment","Degree","Status","Designation","Family_job","Race","Sex","Country"]
```

```python
df.head()
```

```python
for feature in df.columns[:]:
    print(feature, ":", len(df[feature].unique()), "labels")
```

```python
country_map=df["Country"].value_counts().to_dict()
```

```python
df["Country"]=df["Country"].map(country_map)
```

```python
df.head(20)
```

#### Advatages:
1. Easy to use
2. Not increasing feature space
3. information preserve

#### Disadvateges
1. It wil provide same weight if the frequencies are same

## Target Guided Ordinal Encoding: Ordinal

```python
df=pd.read_csv("titanic_train.csv", usecols=["Cabin","Survived"])
```

```python
df.head()
```

```python
df["Cabin"].fillna("Missing", inplace=True) ## Arbitratry Imputation
```

```python
df.head()
```

```python
df["Cabin"]=df["Cabin"].astype(str).str[0]
```

```python
df.head(20)
```

```python
df.Cabin.unique()
```

```python
df.groupby(["Cabin"])["Survived"].mean()
```

```python
ordinal_labels=df.groupby(["Cabin"])["Survived"].mean().sort_values().index
```

```python
ordinal_labels2={k:i for i,k in enumerate(ordinal_labels,1)}
```

```python
ordinal_labels
```

```python
df["Cabin_Ordinalencode"]=df["Cabin"].map(ordinal_labels2)
```

```python
df.head(10)
```

### Mean Encoding:Nominal

```python
mean_encoding=df.groupby(["Cabin"])["Survived"].mean().to_dict()
```

```python
mean_encoding
```

```python
df["mean_nominal_encoding"]=df["Cabin"].map(mean_encoding)
```

```python
df.head()
```

#### Advatages
1. Capture information within the label, therefore rendering more predictive feature
2. Create monotonic relationship b/w target and feature

#### Disadvatage
1. It leads to overfitting

### Encoding with sunbird library

```python
from sunbird.categorical_encoding import frequency_encoding
```

```python
df=pd.read_csv("adult.data", header=None, index_col=None)
columns=[1,3,5,6,7,8,9,13]
df=df[columns]
df.columns=["Employment","Degree","Status","Designation","Family_job","Race","Sex","Country"]
```

```python
for feature in df.columns[:]:
    print(feature, ":", len(df[feature].unique()), "labels")
```

```python
df["Country"].value_counts()
```

```python
frequency_encoding(df, "Country")
```

```python
df.head()
```
